```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

theta = 0.5
mu = lambda y, t: theta * (0. - y)
sigma = 0.1

T_INIT = 0
T_END = 7

N = 1000

DT = (T_END-T_INIT)/N

TS = np.arange(T_INIT,T_END,DT)

Y_INIT = 0

ys = np.zeros(TS.size)
ys[0] = Y_INIT

```

```python
dW = np.random.normal(0,np.sqrt(DT),TS.size)

for t in range(1,TS.size):
    y = ys[t-1]

    ys[t] = y + mu(y,t) * DT + sigma * dW[t]
```

```python
plt.plot(ys)
```
